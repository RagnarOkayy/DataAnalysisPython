Most basic way to mat plot lib is through plt.plot(x,y), if you are running .py files you need plt.show() to see your plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

These line is needed for the rendering of the graphs

```python
x = np.arange(0,10)
x
```

```python
y = 2*x 
y
```

```python
plt.plot(x,y) # adding a ; will remove the object type
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 6) # lower and upper limit for x
plt.ylim(0, 18) # lower and upper limit for y
plt.title('String Title') # will add a title
plt.savefig('myfirstplot.png') # this will save as png
```
