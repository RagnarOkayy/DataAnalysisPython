Most basic way to mat plot lib is through plt.plot(x,y), if you are running .py files you need plt.show() to see your plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

These line is needed for the rendering of the graphs

```python
x = np.arange(0,10)
x
```

```python
y = 2*x 
y
```

```python
plt.plot(x,y) # adding a ; will remove the object type
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0, 6) # lower and upper limit for x
plt.ylim(0, 18) # lower and upper limit for y
plt.title('String Title') # will add a title
plt.savefig('myfirstplot.png') # this will save as png
```

Figure objects for plots

```python
fig = plt.figure(figsize=(10,10)) # this is 10x10 inches
# like 720x720
axes = fig.add_axes([0,0,0.5,1]) # x , y lower left corner of the axes, width, heigh as percentage
# 1, 1 means 100% , 0.5, 0.5 will take only 50% of the canvas
axes.plot(x, y)
```

We can use plt.subplots() to add multiple plots to a figure object, they will be organized as a tupple and the axises will be saved in a numpy array

```python
# The label is labeled (fig, axes)
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots() # when running this you get a set of axis and figures
```

```python
type(fig)
```

```python
type(axes) # a type of axis objects
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) # 1 row and 2 graphs
# axes is a numpy array in this case holding all the axes
axes[0].plot(x,y)
axes[1].plot(a,b) # this way we can plot mulitple axes
```

```python
type(axes) # now we see that axes is an array
```

```python
fig,axes = plt.subplots(nrows = 2, ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

One dimensional columns and fixing spacing issues

```python
fig,axis = plt.subplots(nrows = 3, ncols = 1)
# axis now is one dimensional array so we refer to them with just the first part of the index
#axis[0].plot(x,y)
#axis[1].plot(a,b)

# Since axis now is a list we can loop through it 
for ax in axis:
    ax.plot(x,y)
    
plt.tight_layout() # this will make sure that no set of axis are overlapping over anything else
```

Now lets fix the overlaping manually

```python
fig,axis = plt.subplots(nrows = 2, ncols = 2)
axis[0][0].plot(x,y)
axis[0][1].plot(x,y)
axis[1][0].plot(x,y)
axis[1][1].plot(a,b)
# by default this will overlap
# plt.tight_layout() this will fix it to some degree

fig.subplots_adjust(wspace=0.9, hspace = 0.5) # wspace and hspace are used for addding spaces , representet as fractions based on average axis size
# there are other parameters such as left, right , bottom ,top by default set to none
```

Exporting graphs and changing parameters

```python
fig,axis = plt.subplots(nrows = 2, ncols = 2)
axis[0][0].plot(x,y)


axis[0][1].plot(x,y)


axis[1][0].plot(x,y)
axis[1][0].set_ylabel('Y label 1,0')
axis[1][0].set_xlabel('X label 1,0')
axis[1][0].set_title('Title for graph 1,0')

axis[1][1].plot(a,b)

# you can set the labesl of the graph manually

# we can add a title for the figure
fig.suptitle('This is a figure level title', fontsize=12)
fig.set_figwidth(10)
fig.set_figheight(5)

# Exporting the figure
fig.savefig('newSubplots.png', bbox_inches='tight')

plt.tight_layout()
```
