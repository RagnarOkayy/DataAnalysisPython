# Conditional Filtering

We select rows on a dataframe based on the conditions for a specific column
df["Pop"] > 50 // This will broadcast into a series of boolean values

```python
df = pd.read_csv('/home/zaimi/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/tips.csv')
df.head()
```

```python
df['total_bill'] > 40 # will return a boolean array
```

```python
df[df['total_bill'] > 40] # will return only those bigger than 40
```

```python
df[df['sex'] == 'Male']
```

```python
#df[(df['total_bill'] > 30) and (df['sex'] == 'Male')]
# wont work since the call is ambigous/ and is used for single boolean values
# Use & - and  instead for the comparison
#     | - or 
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')]
```

```python
# filter based on Sat or Sun
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
# this notation is annyoing 
```

```python
# we use isin method
# make a list of options
options = ['Sat', 'Sun']
df['day'].isin(options) # will return a boolean value, like an or operation
```

```python
df[df['day'].isin(options)]
```

.apply() method can be used for custom functions

```python
df.info()
# here can be seen that CC number is an int object
# we cant get the last 4 digits of an integer
```

```python
# what can we do is convvert to string and than select with inverse indexes
str(12345678)[-4:]
# will select from -4 and to the right
```

Pandas have no built in method to convert integer to a string and slice it up. What we have to do is to create a function that does this.

```python
def last_four(num):
    return str(num)[-4:]
```

We use the .apply() function to impact a specifc part of the dataframe, in this case ea column

```python
df['last_four'] = df['CC Number'].apply(last_four) # no need for paranthesees ()
# the function will be applied to every entry in the dataframe
df['last_four']
```

# Functions Dependent on multiple columns
This will require lambda functions

```python
# quick review of lambda exp
# declaration of lambda expressions 
lambda num : num * 2 # declare a value and it returns some operation with tha value
```

Now apply the lambda expression to the column

```python
df['total_bill'].apply(lambda num : num * 2) # here the lambda will return num * 2
```

Now using apply for multple columns

```python
# use the apply method to take 2 columns
# total_bill and tip

def quality(total_bill, tip): # recomanded to call the arguments the same as the columns
    if tip/total_bill > 25:
        return 'Generous'
    else:
        return 'Other'
```

```python
# we need to fromat the table with lambda expressions 
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']),axis=1) # we run the operation through columns, axis is a secondary argument to the lambda function
df['Quality']
```

# Now we make the code run faster using Vectorise, this will both result in cleaner code and faster computation times

```python
                            #function    # arguments
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip']) # this is a simpler way to get multiple columns
df['Quality']
# np vectorize will make the function aware of a numpy array, by itself the function takes single values
# by vectorizing it will make it faster
```
