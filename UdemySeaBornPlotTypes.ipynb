# Seaborn plot types

```python
# Rug Plot, adds a tick for every entry on x, kinda like a barplot
# y is not interpretable
# shows outlieres clearly
```

```python
# Rug Plots can be converted to histograms if we count every entry for x in its range
# We divide the x axis into equal bin, and teh count will show the number of times a datapoint falls inside the bins
# The y axis can represnt a count or a % of the values
```

```python
# KDE stands for Kernel Density Equation - estimation of the normal distribution, in this case the salary
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('notebook/05-Seaborn/dm_office_sales.csv')
df
```

```python
plt.figure(dpi=100)
sns.rugplot(x='salary', data=df, height= 0.5) # one tick for eveery single salary
```

```python
# Do not use distplot
# Use displot
# use histplot
#sns.set(style='darkgrid')
#sns.set(style='whitegrid')
sns.set(style='whitegrid')
sns.displot(data=df, x='salary', bins=20)
```

```python
sns.set(style='ticks') # will just add ticks on each axis
sns.displot(data=df ,x='salary', bins=20)
```

# sns.set(style=) is a high level command that will set the style for everything in the notebook 

```python
sns.set(style= ['whitegrid', 'ticks']) # You can't set multiple styles
sns.displot(data=df, x='salary',bins=20)
```

```python
sns.displot(data= df, x ='salary', bins=20, color='red' ,edgecolor='blue', linewidth=4, ls='--')
# linewidth will increase edge thickness
# ls - linestyle
```

```python
sns.displot(data=df, x='salary', color='red', kde=True, rug=True)
# this can take rug but histplot cant
```

Covering Histograms

```python
sns.histplot(data = df, x='salary', kde=True) # same result as displot()
```

```python
sns.kdeplot(data=df, x = 'salary') # will return just the kde
```

```python
# lets create some examepls
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data= sample_ages, x='age', height=0.5)
```

```python
sns.displot(data=sample_ages, x='age', bins=30, kde=True)
```

```python
# kde plot
sns.kdeplot(data=sample_ages, x='age')
# we have some negative probability so we have to clip it
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100]) # this is a gausian distribution
# clip will set limmits for the grap
```

```python
# bw_adjust will change the sensitivity of the kde curves
# lower values will mean higher sensitivity
sns.kdeplot(data=sample_ages, x='age', bw_adjust=0.1)
```

High values of bandwith will normalize the gaussian distribution leading to normalization

```python
sns.kdeplot(data=sample_ages, x='age', bw_adjust=2, clip=[0,100])
```

Shadding can put empathesis on the graph 

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust= 0.6 ,shade=True)
```
