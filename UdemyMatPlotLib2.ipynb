```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label = 'X vs X')
ax.plot(x,x**2, label= 'X vs X^2')
# 2 different lines, which is which - we can define this using a legend
# with each plot call we add a label
ax.legend(loc = 'lower left' ) # this will acll the legend, we can tell matplot lib where to place the legend
# you can pass also numeric values instead of strings
# or set it manually using a touple with fractional values
ax.legend(loc=(-0.5, 0.5)) # horizontal and vertical position
```
