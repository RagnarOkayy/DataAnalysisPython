# Seaborn trandes customization for ease of use but we can still make matplotlib calls , it acts as an abstrabtion layer for mat plot lib

Scatter plots show the relation between 2 continuous (height, sales, salaries) features, dont confuse with categorical features which represent unique features with no in betweens (color, names, shapes)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'notebook/05-Seaborn/dm_office_sales.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
# sns. followed by plot type
# we can edit before calling sns.
plt.figure(figsize=(12,4), dpi=100) # sns will adjust to the figure parameters
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education') # this will create the scatterplot for you
# hue='column' it means collor the points based on this particular categorys
```

Lets explore the SeaBorn parameters

```python
# you can also provide a continuous column
plt.figure(figsize=(12,4),dpi=100)
sns.scatterplot(x = 'salary', y ='sales', data=df, hue='salary') # creates a gradiant for you based on the x axis
```

```python
# the colors can be changed using pallets
plt.figure(figsize=(12,4), dpi=100) # sns will adjust to the figure parameters
# this is a categorical columns so we need a categorical pallet too
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2')
```

```python
# Scaling with size
# the colors can be changed using pallets
plt.figure(figsize=(12,4), dpi=100) 
sns.scatterplot(x='salary', y='sales', data=df, size='work experience', s = 200, alpha = 0.5) # the more work exp the bigger teh size
# s will act like a scaler for the size
# alpha mean transparency , 1 - opaque (not transparent), 0 - transparent 
```

```python
plt.figure(figsize=(12,4), dpi=100) 
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', s = 100, hue='level of education', palette='Dark2') # style will change the icon based on the level of education, should be categorical
# we can combine style and hue to make the columns more redeable
```
